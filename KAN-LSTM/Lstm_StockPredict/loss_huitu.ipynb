```python
import matplotlib.pyplot as plt

# 读取文本文件中的损失数据
file_path = "lstm_loss_log.txt"
epochs = []
losses = []
with open(file_path, 'r') as file:
    for line in file:
        if ', Loss: ' in line:
            epoch, loss = line.strip().split(', Loss: ')
            epochs.append(int(epoch.split()[1]))
            losses.append(float(loss))

# 绘制损失曲线
plt.figure(figsize=(10, 5))
plt.plot(epochs, losses, marker='o')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Over Epochs')
plt.grid(True)
plt.show()
```

```python
import re
import matplotlib.pyplot as plt

# 读取文本文件中的损失数据
file_path = "kan_lstm_loss_log.txt"
epochs = []
losses = []
with open(file_path, 'r') as file:
    for line in file:
        match = re.search(r'Epoch (\d+)/\d+, Loss: (.+)', line)
        if match:
            epoch = int(match.group(1))
            loss = float(match.group(2))
            epochs.append(epoch)
            losses.append(loss)

# 绘制损失曲线
plt.figure(figsize=(10, 5))
plt.plot(epochs, losses, marker='o')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Over Epochs')
plt.grid(True)
plt.show()
```
